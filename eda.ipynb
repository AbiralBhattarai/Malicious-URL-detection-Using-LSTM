# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint
```

```python
df = pd.read_csv('dataset.csv')
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
print(df)
```

```python
df['type'] = df['type'].replace('benign', 0)
df['type'] = df['type'].replace('malicious', 1)
df.to_csv('dataset1.csv')
```

```python
data = pd.read_csv("balanced dataset.csv")
datagrp = data.groupby('type')
benign_data = datagrp.get_group(0)
malicious_data = datagrp.get_group(1)
```

```python
type_count = df['type'].value_counts()
print(type_count)

```

```python
sns.barplot(data = type_count )
```

```python
count = {
'benign' : [benign_data['type'].count()],
'malicious' : [malicious_data['type'].count()]
}
count = pd.DataFrame(count,index=['count'])
```

```python
sns.barplot(data = count)
```
